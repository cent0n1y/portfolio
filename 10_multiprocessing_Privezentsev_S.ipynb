````python
%%file analyze.py
from collections import Counter

def analyze_symbols(f_name, priority_res):
    count = Counter()
    with open(f_name, 'r') as f:
        for line in f:
            count.update(line.lower())
    priority_res.put((f_name, dict(count)))
````

````python
%%time 
import analyze
import multiprocessing as mp

priority_res = mp.Queue()
analyze.analyze_symbols("/Users/apple/Desktop/УНИК/TOBD/Dostoevskiy Fedor. Igrok - BooksCafe.Net.txt",encoding = 'cp1251', priority_res)
analyze.analyze_symbols("/Users/apple/Desktop/УНИК/TOBD/Dostoevskiy Fedor. Prestuplenie i nakazanie - BooksCafe.Net.txt'", priority_res)

while not priority_res.empty():
    f_name, count_symb = priority_res.get()
    print(f"Количество символов, встречающихся в файле: {f_name}:\n{count_symb}")
    
print("Привезенцев Семен")    
````

2. Напишите функцию, которая принимает на вход название файла, созданного в результате решения задачи 1, считает среднее значение количества шагов для каждого тэга и возвращает результат в виде словаря.

````python
import pandas as pd
def average_steps(f_name):
    df = pd.read_csv(f_name)
    all_tag = {}
    av_tag = {}
    
    for ind, data in df.iterrows():
        for tag in eval(data.tags):#data.tags - str, приводим ее к типу list 
            if tag not in all_tag:
                all_tag[tag] = {'steps': 0, 'recipes': 0}
                
            all_tag[tag]['steps'] += data.n_steps
            all_tag[tag]['recipes'] += 1
    
    for tag, totals in all_tag.items():
            av_tag[tag] = round(totals['steps'] / totals['recipes'], 4)
    return av_tag
print("Привезенцев Семен")
````

````python
%%file average.py

import multiprocessing as mp
import pandas as pd
def average_all_steps():
    files = [f'id_tag_nsteps_{i}.csv' for i in range(8)]
    res = list()
    pool = mp.Pool(processes=len(files)) 
    results = pool.map(file_process, files) 
    pool.close()
    pool.join() 
    return combine_results(res)

def file_process(f_name):
    df = pd.read_csv(f_name)
    all_tag = {}
    tags = df.tags
    
    for ind, data in df.iterrows():
        for tag in eval(data.tags):
            if tag not in all_tag:
                all_tag[tag] = {'steps': 0, 'recipes': 0}
                
            all_tag[tag]['steps'] += data.n_steps
            all_tag[tag]['recipes'] += 1

    return all_tag

def combine_results(res):
    av_total = {}
    res = {}

    for dic in results:
        for tag, v in dic.items():
            if tag not in av_total:
                av_total[tag] = v
            else:
                av_total[tag]['steps'] += v['steps']
                av_total[tag]['recipes'] += v['recipes']
                
    for tag, totals in av_total.items():
        res[tag] = round(totals['steps'] / totals['recipes'], 4)
    return res
print('Привезенцев Семен')
````
