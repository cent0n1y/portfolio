```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

G.add_nodes_from([1, 2, 3, 4, 5])

G.add_edges_from([(1, 2),(1,3),(2, 3), (3, 4), (4, 5), (5, 1)])

pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True, arrows=True, node_size=700, node_color="skyblue", font_size=10, font_color="black", font_weight="bold", width=1.5, edge_color="gray", style="dashed")

plt.show()
```
